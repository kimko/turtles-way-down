```python
f_median_carapace_by_age = natives[natives.Gender == 'f'][['Carapace','new_annuli']].groupby('new_annuli').mean()
m_median_carapace_by_age = natives[natives.Gender == 'm'][['Carapace','new_annuli']].groupby('new_annuli').mean()
median_carapace_by_age = pd.concat([f_median_carapace_by_age,m_median_carapace_by_age],axis=1)
median_carapace_by_age.columns = ['f_median_Carapace','m_median_Carapace']
median_carapace_by_age.reset_index(inplace=True)
display(median_carapace_by_age)

natives = pd.merge(natives,median_carapace_by_age,how='outer',on='new_annuli')
```
